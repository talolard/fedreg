```python
import scrapy
from scrapy.crawler import CrawlerProcess
import json
```

```python

class BlogSpider(scrapy.Spider):
    name = 'blogspider'
    start_urls = ['https://www.state.gov/{yr}-TIAS/'.format(yr=yr) for yr in range(1981,2020)]
    custom_settings = {
    'FEED_FORMAT': 'json',
    'FEED_URI': 'test1.json'
    }


    def parse(self, response):
        for title in response.css('.collection-result__link'):
            res = {'title': title.css('a ::text').get()}
            print(res)
            yield(res)

        for next_page in response.css('a.next'):
            yield response.follow(next_page, self.parse)
```

```python
process = CrawlerProcess()
process.crawl(BlogSpider)
results = process.start()
```

```python
d= json.load(open('./test1.json'),)
```

```python
import pandas as pd
D= pd.DataFrame(d)
```

```python

D.title =D.title.str.strip()
```

```python
import re
reg = re.compile('(?P<country>.+?) \((?P<code>[.\dA*-]+) *\)[ –-]+(?P<name>.+)')
reg.search(d[0]['title']).groupdict()
```

```python
def to_dict(item):
    txt = item['title'].strip()
    try:
        return reg.search(txt).groupdict()
    except:
        print(txt)
Z =list(filter(lambda x:x,map(to_dict,d)))
```

```python
def make_trigrams(text):
    grams = text.lower().split(' ')
    trigrams = list(zip(grams,grams[1:],grams[2:]))
    return trigrams
    
```

```python
grams =sum(map(lambda x:make_trigrams(x['name']),Z),[])
```

```python
V = pd.Series(grams).value_counts()
V = V[V>30]
phrases = V.index.map(lambda x: ' '.join(x)).tolist()
json.dump(phrases,open('phrases.json','w'))
```
